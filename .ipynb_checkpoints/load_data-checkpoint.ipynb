```python
import pandas as pd
import os
```

```python
def load_raw_data():
    """
    Load the raw CSV dataset from disk.
    """
    base_dir = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
    file_path = os.path.join(base_dir, "data", "GlobalWeatherRepository.csv")
    df = pd.read_csv(filepath)
    return df


def standardize_columns(df):
    """
    Standardize column names:
    - Remove leading/trailing spaces
    - Convert to lowercase
    - Replace spaces and hyphens with underscores

    This ensures consistency when referencing columns later.
    """
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace("-", "_", regex=False)
    )
    return df


def parse_datetime(df, datetime_col="lastupdated"):
    """
    Convert the 'lastupdated' column to datetime format.

    Rows where parsing fails are removed to maintain
    a clean time-series structure.
    """
    df[datetime_col] = pd.to_datetime(df[datetime_col], errors="coerce")

    # Drop rows where datetime conversion failed
    df = df.dropna(subset=[datetime_col])

    return df


def drop_duplicates(df):
    """
    Remove duplicate rows to prevent repeated
    time-series entries.
    """
    df = df.drop_duplicates().reset_index(drop=True)
    return df


def create_timeseries_index(df):
    """
    Create a multi-index for proper time-series analysis.

    Index structure:
        (city, country, lastupdated)
    or
        (city, lastupdated)

    This enables grouped time-series modeling per city.
    """
    if "country" in df.columns:
        df = df.set_index(["city", "country", "lastupdated"])
    else:
        df = df.set_index(["city", "lastupdated"])

    # Sort to ensure chronological order
    df = df.sort_index()

    return df


def load_clean_data():
    """
    Complete data loading pipeline:
        1. Load raw CSV
        2. Standardize column names
        3. Parse datetime
        4. Remove duplicates
        5. Create time-series index

    Returns a clean dataframe ready for EDA and modeling.
    """
    df = load_raw_data()
    df = standardize_columns(df)
    df = parse_datetime(df)
    df = drop_duplicates(df)
    df = create_timeseries_index(df)

    return df
```
