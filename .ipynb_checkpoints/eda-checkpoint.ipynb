```python
# Global Trend – Average Temperature Over Time

df_reset = df_out.reset_index()

# Ensure datetime column is proper
df_reset["last_updated_epoch"] = pd.to_datetime(df_reset["last_updated_epoch"])

# Group by time (daily average across all locations)
global_temp = (
    df_reset
    .groupby("last_updated_epoch")["temperature_celsius"]
    .mean()
)

import matplotlib.pyplot as plt

plt.figure()
plt.plot(global_temp.index, global_temp.values)
plt.title("Global Average Temperature Over Time")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
